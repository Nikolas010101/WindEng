```python
from matplotlib import pyplot as plt
from io import StringIO
import numpy as np, csv
```

```python
def parse_dat(path) -> tuple[str, np.ndarray, np.ndarray]:
    with open(path, "r") as file:
        dat_data = file.read()
    file = StringIO(dat_data[dat_data.find("\n")+1:].replace("     ", " "))
    reader = csv.reader(file, delimiter=" ")
    coords = np.array([row for row in reader], dtype=np.float64)
    x, y = coords[:, 0], coords[:, 1]
    return dat_data[:dat_data.find("\n")], x, y
```

```python
def plot_airfoil(turb_name, x, y) -> None:
    fig = plt.gcf()
    fig.set_size_inches(20, 3)
    axes = plt.gca()
    axes.set_xticks(np.arange(-0.5, 1.5, 0.1))
    axes.set_yticks(np.arange(-0.5, 0.15, 0.02))
    fig.set_dpi(300)
    plt.title(turb_name)
    plt.plot(x, y)
    plt.grid()
    plt.show()
```

```python
plot_airfoil(*parse_dat("4415.dat"))
```
